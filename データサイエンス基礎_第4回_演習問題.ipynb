## B1: Shuffle PULS


### 問題文

整数列 $A=(A_1, A_2, A_3, ... , A_n)$ が与えられます。

$A$ を並べ替えた列を $B=(B_1, B_2,B_3, ... ,B_n)$ とします。

このとき、$B_1 + B_2 + ... + B_{n-1} = B_n$ を満たすようにできるか判定してください。

### 制約

- 入力は全て整数

- $3 ≤ N ≤ 10$
- $1 ≤ A_i ≤ 100$

### 入力形式

```
N
A1 A2 ... AN
```

### 出力形式

$B_1 + B_2 + ... + B_{n-1} = B_n$ を満たすようにできるなら`Yes`、そうでないなら`No`と出力せよ。


### 入力例1

```
3
3 8 5
```

$A = (3, 8, 5)$ です。

$B = (3, 5, 8)$と並べ替えることで、$B_1+B_2 = B_3$を達成できます。

### 出力例1

```
Yes
```

### 入力例2

```
5
1 1 5 1 1
```

### 出力例2

```
No
```

### 入力例3

```
10
28 8 73 58 12 82 76 443 74 32
```

### 出力例3

```
Yes
```
