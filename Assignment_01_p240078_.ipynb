## 課題 1 (最適化，発展課題)

```python
import sys

# 課題の数を定義
# 可変にする
num_of_kadai = int(input("課題数を入力してください: "))

# 入力された点数を格納するための空のリスト

scores = []

# forを使って、定義した課題の数だけ点数の入力を繰り返す
for i in range(num_of_kadai):
 #while文で不正な入力を弾く
    while True:
        try:
            # 入力を受け取る
            raw_input = input(f"課題{i+1}の点数を入力してください: ")
            
            # 何も入力されなかった場合、ループの最初に戻って再入力を促す
            if raw_input == "":
                print("何も入力されていません。点数を入力してください。")
                continue # 以降の処理をスキップし、次のループに進む
            
            # 整数に変換する
            score = int(raw_input)
            
            # 点数が0未満、または100より大きい場合は終了する
            if not (0 <= score <= 100):
                print("エラー: 点数は0以上100以下の範囲で入力してください。")
                sys.exit()
            
            
            # 正しい点数が入力されたら、whileループを抜ける
            break

        except ValueError:
            # int()での変換に失敗した場合
            print("エラー: 半角数字で点数を入力してください。")

    # whileループを正常に抜けたら、正しい点数が配列に入る
    scores.append(score)

avg = sum(scores) / len(scores)
print(f"平均点: {avg}")
```

## 課題2 (最適化，発展課題)

```python
import math
import sys

# 課題の数を定義
# 可変にする

num_of_kadai = int(input("課題数を入力してください: "))

# 入力された点数を格納するための空のリスト

scores = []

# forを使って、定義した課題の数だけ点数の入力を繰り返す
for i in range(num_of_kadai):
    #while文で不正な入力を弾く
    while True:
        try:
            # 入力を受け取る
            raw_input = input(f"課題{i+1}の点数を入力してください: ")
            
            # 何も入力されなかった場合、ループの最初に戻って再入力を促す
            if raw_input == "":
                print("何も入力されていません。点数を入力してください。")
                continue # 以降の処理をスキップし、次のループに進む
            
            # 整数に変換する
            score = int(raw_input)
            
            # 点数が0未満、または100より大きい場合は終了する
            if not (0 <= score <= 100):
                print("エラー: 点数は0以上100以下の範囲で入力してください。")
                sys.exit()
            
            
            # 正しい点数が入力されたら、whileループを抜ける
            break

        except ValueError:
            # int()での変換に失敗した場合
            print("エラー: 半角数字で点数を入力してください。")

    # whileループを正常に抜けたら、正しい点数が配列に入る
    scores.append(score)

    avg = sum(scores) / len(scores)
    variance_sum = sum([(s - avg)**2 for s in scores])
    var = variance_sum / len(scores)
    std = math.sqrt(var) 
print(f"平均点: {avg}")
print(f"分散: {var}")
print(f"標準偏差: {std}")
```
