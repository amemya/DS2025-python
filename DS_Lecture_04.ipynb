## 3. `break` / `continue`
---

while æ–‡ã¨ for æ–‡ã‚’åˆ¶å¾¡ã™ã‚‹å‘½ä»¤ã¨ã—ã¦ã€**break** ã¨ **continue** ãŒã‚ã‚Šã¾ã™ã€‚

#### break
---
break ã¯ãƒ«ãƒ¼ãƒ—ã‚’é€”ä¸­ã§æŠœã‘ã‚‹ãŸã‚ã®å‘½ä»¤ã§ã™ã€‚

````python
# break ãŒãªã‘ã‚Œã°ã“ã®ãƒ«ãƒ¼ãƒ—ã¯ i == 4 ã¾ã§ç¹°ã‚Šè¿”ã™
for i in range(5):
    if i == 3:
        print("ã¬ã‘ã‚‹")
        break   # i == 3 ã®æ™‚ç‚¹ã§ãƒ«ãƒ¼ãƒ—ã‹ã‚‰æŠœã‘ã‚‹
        
    print(i)
    
print("çµ‚äº†")
````

#### continue
---

## 4. `for-else` / `while-else`

Python ã§ã¯ for æ–‡ã¨ while æ–‡ã« else ç¯€ã‚’æ›¸ãã“ã¨ãŒã§ãã¾ã™ã€‚

ã“ã‚Œã¯ã€ãƒ«ãƒ¼ãƒ—ãŒ break æ–‡ã§çµ‚äº†ã—ãŸå ´åˆã¯**å®Ÿè¡Œã•ã‚Œãš**ã€

ãã†ã§ãªã„ã¨ãï¼ˆãƒ«ãƒ¼ãƒ—ãŒæœ€å¾Œã¾ã§å›ã£ãŸã¨ãï¼‰ã«å®Ÿè¡Œã•ã‚Œã¾ã™ã€‚

#### ä¾‹ï¼šbreakã§çµ‚äº†ã—ãªã„foræ–‡ï¼ˆelseãŒå®Ÿè¡Œã•ã‚Œã‚‹ï¼‰

````python
for i in range(5):
    print(i)
else:
    print("ãƒ«ãƒ¼ãƒ—ãŒæœ€å¾Œã¾ã§å›ã‚Šã¾ã—ãŸã€‚")

print("çµ‚äº†")
````

#### ä¾‹ï¼šbreakã§çµ‚äº†ã™ã‚‹foræ–‡ï¼ˆelseãŒå®Ÿè¡Œã•ã‚Œãªã„ï¼‰

````python
for i in range(5):
    if i == 3:
        print("ã¬ã‘ã‚‹")
        break  # i == 3 ã®æ™‚ç‚¹ã§ãƒ«ãƒ¼ãƒ—ã‹ã‚‰æŠœã‘ã‚‹

    print(i)
else:
    print("ãƒ«ãƒ¼ãƒ—ãŒæœ€å¾Œã¾ã§å›ã‚Šã¾ã—ãŸã€‚")

print("çµ‚äº†")
````

## ãƒªã‚¹ãƒˆ

**ãƒªã‚¹ãƒˆï¼ˆé…åˆ—ã€list å‹ï¼‰** ã¯ãƒ‡ãƒ¼ã‚¿ã®åˆ—ã‚’æ‰±ã†ã“ã¨ãŒã§ãã‚‹éå¸¸ã«é‡è¦ãªæ©Ÿèƒ½ã§ã™ã€‚ä¾‹ãˆã°ã€ä»¥ä¸‹ã®ã‚ˆã†ãªã‚³ãƒ¼ãƒ‰ã‚’æ›¸ãã“ã¨ãŒã§ãã¾ã™ã€‚

````python
# ãƒªã‚¹ãƒˆ [3, 1, 4, 1, 5] ã‚’ä½œã‚Š, a ã«ä»£å…¥ã™ã‚‹
a = [3, 1, 4, 1, 5]

# a ã® 0 ç•ªç›®ã®è¦ç´ ã§ã‚ã‚‹ 3 ã‚’å‡ºåŠ›
print(a[0])

# a ã® 1 ç•ªç›®ã®è¦ç´ ã§ã‚ã‚‹ 1 ã‚’å‡ºåŠ›
print(a[1])

# a ã® 2 ç•ªç›®ã®è¦ç´  (4) ã‚’ 7 ã«å¤‰æ›´
a[2] = 7

# a ã®è¦ç´ ã‚’é †ã«å‡ºåŠ›
for x in a:
    print(x)
````

````python
a = [3, 1, 4, 1, 5]
print(len(a))
````

### ãƒªã‚¹ãƒˆã® $i$ ç•ªç›®ã®è¦ç´ 
- ãƒªã‚¹ãƒˆ $a$ ã¨æ•´æ•° $i$ ã«å¯¾ã—ã€`a[i]` ã§ $a$ ã® $i$ ç•ªç›®ã®è¦ç´ ã‚’å–å¾—ã§ãã¾ã™ã€‚

````python
# ãƒªã‚¹ãƒˆ [3, 1, 4, 1, 5] ã‚’ä½œã‚Šã€a ã«ä»£å…¥ã™ã‚‹
a = [3, 1, 4, 1, 5]

# a ã®é•·ã• 5 ã‚’å‡ºåŠ›ã™ã‚‹
print(len(a))

# a ã® 2 ç•ªç›®ã®è¦ç´  (4) ã‚’ 7 ã«å¤‰æ›´ã™ã‚‹
a[2] = 7

# range(len(a)) ã¨æ›¸ã‘ã° a ã®æ·»å­— (0, 1, 2, 3, 4) ã‚’é †ã«åˆ—æŒ™ã§ãã‚‹
for i in range(len(a)):
    # a[i] ã§ a ã® i ç•ªç›®ã®è¦ç´ ã‚’å–å¾—ã™ã‚‹
    print(i, ":", a[i])
````

- ã“ã® $i$ ã®ã“ã¨ï¼ˆãƒªã‚¹ãƒˆã®ä½•ç•ªç›®ã®è¦ç´ ã‹ã‚’è¡¨ã™ç•ªå·ï¼‰ã‚’ æ·»å­—ï¼ˆããˆã˜ï¼‰ ã¨è¨€ã„ã¾ã™ã€‚

- `a[i] = x` ã§ $a$ ã® $i$ ç•ªç›®ã®è¦ç´ ã‚’ $x$ ã«å¤‰æ›´ã§ãã¾ã™ã€‚

## for æ–‡ã§ãƒªã‚¹ãƒˆã®è¦ç´ ã‚’åˆ—æŒ™ã™ã‚‹

- ãƒªã‚¹ãƒˆ $a$ ã«å¯¾ã—ã€`for x in a:` ã§ $a$ ã®è¦ç´ ã‚’é †ã« $x$ ã«ä»£å…¥ã—ã¦ç¹°ã‚Šè¿”ã—å‡¦ç†ã‚’è¡Œã†ã“ã¨ãŒã§ãã¾ã™ã€‚

### ãƒªã‚¹ãƒˆã®å¥½ããªä½ç½®ã«è¦ç´ ã‚’è¿½åŠ ãƒ»å‰Šé™¤ã™ã‚‹

* `a.insert(i, x)` ã§ `a[i-1]` ã¨ `a[i]` ã®é–“ã« `x` ã‚’æŒ¿å…¥ã€‚
* `a.pop(i)` ã§ `a[i]` ã‚’å–å¾—ã—ãªãŒã‚‰å‰Šé™¤ã€‚

### ãƒªã‚¹ãƒˆã‚’ã‚½ãƒ¼ãƒˆãƒ»åè»¢ã™ã‚‹

### ãƒªã‚¹ãƒˆã‚’è¤‡è£½ã™ã‚‹
---

ãƒªã‚¹ãƒˆ 
$a$ ã«å¯¾ã— `b = a` ã¨ã™ã‚‹ã¨ã€ãƒªã‚¹ãƒˆã¯è¤‡è£½ã•ã‚Œãšã€$a$ ã®æŒ‡ã™ãƒªã‚¹ãƒˆã¨ $b$ ã®æŒ‡ã™ãƒªã‚¹ãƒˆã¯åŒä¸€ã®ã‚‚ã®ã«ãªã‚Šã¾ã™ã€‚

ã“ã‚Œã‚’å›é¿ã™ã‚‹ã«ã¯ã€`b = a.copy()` ã¨æ›¸ã„ã¦ã€ãƒªã‚¹ãƒˆã‚’è¤‡è£½ã™ã‚‹å¿…è¦ãŒã‚ã‚Šã¾ã™ã€‚
