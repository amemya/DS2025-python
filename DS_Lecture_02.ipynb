### 3-3. 変数について

以下のプログラムは何の計算するプログラムでしょうか？

````python
print(800*2 + 250*5 + 120*8)
````

わからないですね。ただ、こう書かれていたらどうでしょうか？

````python
cake = 800
chocolate = 250
gummy = 120
price = cake * 2 + chocolate * 5 + gummy * 8
print(price)
````

これは何の計算をしているのか、わかりますね。

ケーキ2個とチョコ5個とグミ8個の値段を計算していそうです。

これが変数を使うメリットの一つです。**後から見ても、意味がすぐに理解できます。**

一方で、最初から数値が直接書かれていて、「この数字は何のためにあるのだろう？」と思うようなものを**マジックナンバー**と呼びます。
マジックナンバーは、できるだけ使わないようにしましょう。

将来、大きなシステムを作るときには、複数人でプログラムを開発することになります。
そのとき、誰が見てもわかる**読みやすいコード（可読性の高いコード）**を書くことがとても大切です。

変数を使うもう一つの大きなメリットは、**値の変更がしやすい**ことです。

たとえば、プログラムの中で同じ数字を何度も使っている場合、もしその数字を変更したくなったときに、すべての箇所を探して書き換えるのは大変ですよね。
でも、変数を使っていれば、その変数の値を一か所変えるだけで、プログラム全体に反映されます。

このように、変数を使うことでプログラムの**保守（メンテナンス）**がしやすくなるのです。

さらに、同じ意味で同じ数字を使いたいときにも、変数はとても便利です。
たとえば「税率」や「スピードの上限」など、意味のある数字を変数として定義しておくと、再利用性が高く、他の場所でも同じ変数を使うことで整合性を保つことができます。

>**💡 TIPS:**
>
>もう一つ大事なのは、「次の日の自分を信用しない」ことです。
一晩経つと、**自分で書いたはずの数字の意味がわからなくなること**はよくあります。
だからこそ、後で見ても理解しやすいように、変数を使って意味のわかるコードを書くようにしましょう。

- **変数のメリット**：値に名前をつけることで、再利用・保守・可読性が向上
- **使い方**：`名前 = 値` で代入

#### 変数の例

````python
score_math = 80
score_english = 72
average = (score_math + score_english) / 2

print("数学:", score_math)
print("英語:", score_english)
print("平均:", average)
````

### 3-4. 型（type）について

変数には **「型」** があります。  
その変数に**どんな種類の値が入っているか**を表すものです。

型にはいくつかの種類があります。
代表的なものをいくつか上げておきます。

- **整数（int）**：例）1, 0, -3
- **実数（float）**：例）3.14, -0.5
- **文字列（str）**：例）"hello", 'こんにちは'

#### 動的型付けについて

Pythonでは、**代入した値によって自動的に型が決まります。**  
この仕組みを **「動的型付け」** と呼びます。

```python
x = 10      # 整数型（int）
x = "Hello" # 文字列型（str）
```
このように、同じ変数でもあとから違う型の値を代入できるのがPythonの特徴です。

動的型付けはとても便利ですが、
**思わぬエラー**の原因になることもあります。

たとえば、変数が自分の想定とは違う型になってしまっていることがあります。

````python
x = "10"
print(x + 5)  # エラー：「文字列」と「数値」は足せない
````

内部的な型の種類は`type`関数を使って調べることができるので、型でエラーが起きた場合は`type`を使って調べます。

````python
# 型の例と平均の計算
a = 5          # int
print(type(a), a)
````

````python
b = 2.0          # float
print(type(b), b)
````

````python
c = a / b      # float
print(type(c), c)
````

````python
d = "Python"  # str
print(type(d), d)
````

````python
e = input()  #str
print(type(e), e)
````

#### 型の変換

`input()`で入力された文字は必ず文字列として入力されます。
`input()`で受け取った値は文字列（`str`型）なので、そのままでは数値の計算ができません。
計算などを行いたい場合には、明示的に型を変換する必要があります。


**代表的な変換関数**

| 関数        | 変換前    | 変換後        | 例                      |
| --------- | ------ | ---------- | ---------------------- |
| `int()`   | 文字列や小数 | 整数（int型）   | `int("10") → 10`       |
| `float()` | 文字列や整数 | 小数（float型） | `float("3.14") → 3.14` |
| `str()`   | 数値など   | 文字列（str型）  | `str(123) → "123"`     |


#### 例1：文字列 → 整数

````python
num_str = input("好きな数字を入力してください: ")  # 入力例: "5"
num = int(num_str)  # 整数に変換
print(num * 2)  # 10 と表示される
````

#### 例2：文字列 → 小数

````python
num = float(input("身長を入力してください（例: 160.5）: "))
print("入力された身長は", num, "cmです。")
````

#### 例3：数値 → 文字列

````python
age = 20
text = "私は" + str(age) + "歳です。"
print(text)
````
